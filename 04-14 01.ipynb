```python
import numpy as np

a = np.array([1, 2, 3])              # shape (3,)
B = np.array([[4, 5], [6, 7], [8, 9]])  # shape (3, 2)

result = a @ B
print(result)   # [40 46]
print(result.shape)  # (2,)
```

```python
def do_something(x, callback=None):
    x += 1
    if callback:
        x = callback(x)
    return x

print(do_something(10))          # 11
print(do_something(10, lambda x: x*2))  # 22
```

```python
class MyDataset:
    def __init__(self, data, transform=None):
        self.data = data
        self.transform = transform  # 선택적 함수 저장

    def __getitem__(self, idx):
        sample = self.data[idx]
        if self.transform:          # 선택적 함수가 있으면 적용
            sample = self.transform(sample)
        return sample

def add_noise(x):
    return x + 1

dataset1 = MyDataset([10, 20, 30])             # transform 없음
dataset2 = MyDataset([10, 20, 30], add_noise)  # transform 있음

print(dataset1[0])  # 10
print(dataset2[0])  # 11
```

```python
class MyList:
    def __init__(self, data):
        self.data = data

    def __getitem__(self, index):
        print(f"__getitem__ 호출! index={index}")
        return self.data[index]

# 객체 생성
lst = MyList([10, 20, 30])

# 객체[] =>  __getitem__
# 인덱스로 접근 =>  __getitem__ 자동 호
print(lst[0])  # __getitem__ 호출! index=0 → 10
print(lst[1])  # __getitem__ 호출! index=1 → 20
```

```python
class Adder:
    def __init__(self, n):
        self.n = n

    def __call__(self, x):
        return x + self.n

add5 = Adder(5)   # 객체 생성
print(add5(10))   # __call__ 자동 실행 → 15
```
