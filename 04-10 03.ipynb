```python
# user profile
users_num=10

movies_num=300

# 10행 300열짜리 0/1 matrix형태
# 0~1중에서 랜덤하게 수를 뽑아 10행 300열의 행렬만듬 (0: 해당영화보지않음 1: 해당영화 시청함)
# 시청기록데이터가 없으므로 랜덤하게(0~1) 데이터 설정함
users_movie=np.random.randint(2, size=(users_num, movies_num))
users_movie=pd.DataFrame(users_movie, columns=item_profile['title'])
users_movie
```

```python
# 영화의 정보들(영화속성) + 유저의 시청기록 => 행렬 곱 (dot product)
# 유저 취향 벡터 뽑으려고
def user_profile(item_profile, users_movie):
    movie_columns=item_profile.columns.drop('title')
    # 유저의 수만큼 0으로 채움(비어있게 만듬- 초기화)
    # 타입문제로 0.0
    user_profiles=pd.DataFrame(0.0, index=range(len(users_movie)), columns=movie_columns)

    # 시청한 영화 골라내기
    for user_id in range(len(users_movie)):
        w_movies = users_movie.iloc[user_id] == 1 #시청기록 중 숫자가 1인것만 true로 표시됨
        w_titles = users_movie.columns[w_movies] #시청한 영화제목만 뽑아온다

        # item profile에서 유저가 시청한 영화만 뽑아온다(시청한 영화에 대한 정보들이 담겨져있다)
        w_profiles = item_profile[item_profile['title'].isin(w_titles)]

        # 시청한 영화가 없는 경우
        if w_profiles.empty:
            user_profile1 = pd.Series(0, index=movie_columns)
        # 영화를 봤으면 평균계산(영화특징들)
        else:
            user_profile1 = w_profiles[movie_columns].mean()
            user_profile1 = user_profile1.fillna(0)
            
        user_profiles.iloc[user_id] = user_profile1
    return user_profiles.fillna(0)

user1=user_profile(item_profile, users_movie)
user1
```

```python
# 코사인 유사도 계산
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# 유저의 취향을 알았으니, 모든 영화와 대조해서 가장 취향에 맞는 5개의 영화를 골라줌
def movie_recommend(item_profile, user_profiles, top_n=5):
    recommend={}

    movie_columns=item_profile.columns.drop('title')
    item_features=item_profile[movie_columns].fillna(0)

    for user_id in range(len(user_profiles)):
        # 1차원 -> 2차원으로 만듬(1행은 고정 열은 자동으로 맞춰줌)
        user_profile=user_profiles.iloc[user_id].values.reshape(1,-1)
        user_profile=np.nan_to_num(user_profile)

        # 유저의 취향벡터와 300개 영화벡터 사이의 각도를 잰다 (0~1)
        similarity=cosine_similarity(user_profile, item_features)

        # 점수가 낮은 순서대로 인덱스 번호 준 다음에 뒤에서 5개를 자른다.
        # 순서를 뒤집어서 가장 높은 점수가 맨 앞으로 오게 만든다.
        top_similarity=np.argsort(similarity[0][-top_n:][::-1])

        # 추천영화 담을 빈 리스트
        recom_movies=[]

        for i in top_similarity:
            # 유사도가 높은 5개를 반복 순회하면서 해당 인덱스의 영화제목을 뽑는다.
            movie_title=item_profile.iloc[i]['title']
            similarity_score=similarity[0][i]
            recom_movies.append({
                'title':movie_title,
                'similarity_score':similarity_score})
        # 딕셔너리에 저장 dic['age']=20 dic{'age':20}
        recommend[f'{user_id}']=recom_movies

    return recommend

recom=movie_recommend(item_profile, user_profiles)
recom
```
