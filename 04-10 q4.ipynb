```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris

df=load_iris()
df
```

```python
features=df.data
label=df.target
```

```python
X_train, X_test, y_train, y_test=train_test_split(features, label,test_size=0.2, random_state=1)
```

```python
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.decomposition import PCA

pca=PCA(n_components=2)
```

```python
X_train_pca=pca.fit_transform(X_train)

X_test_pca=pca.transform(X_test)
```

```python
print(pca.explained_variance_ratio_)
```

```python
from sklearn.cluster import KMeans

k_means=KMeans(n_clusters=3)

k_means.fit(X_train_pca)
```

```python
k_means.labels_

centers=k_means.cluster_centers_
```

```python
from matplotlib.colors import ListedColormap

color_bold=ListedColormap(['red','blue','green'])
color_light=ListedColormap(['#FFAAAA','#42e3f5','#42f584'])

plt.figure(figsize=(10,8))

for i in range(3):
    clusters=k_means.labels_ == i
    plt.plot(X_train_pca[clusters,0], X_train_pca[clusters, 1], '.', c=color_light(i), markersize=10)
    plt.plot(centers[i,0], centers[i,1], 'o', c=color_bold(i), markersize=10)
```
