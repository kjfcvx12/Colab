```python
d1=np.array([[1,2],[3,4]])
print(type(d1))

# 파이토치에서 텐서, 넘파이의 array 서로 변환이 가능하다.

# 넘파이의 ndarray를 tensor객체로 변환하겠다
d2=torch.from_numpy(d1)
print(type(d2))

#텐서객체를 넘파이의 ndarray로 변환하겠다.
d3=d2.numpy()
print(type(d3))
```

```python
d1=torch.IntTensor([[1,2,3],
                    [4,5,6],
                    [7,8,9]])

print(d1[:,1])

# boolean indexing
d2=d1>5
print(d2)
print(d1[d2]) # d1[5보다 큰값]
print(d1[(d1>7)&(d1<10)])
```

```python
d1=torch.FloatTensor([1,2,3,4])
# where : 조건에 맞는 인덱스 번호 리턴
w1=torch.where(d1<4)
print(w1)
print(d1.max())
print(d1.min())
print(d1.mean())
print(d1.sum())
```

```python
# *******squeeze : 차원 압축 -> 차원이 1인 축을 제거함**********
d1=torch.FloatTensor([[1],
                      [2],
                      [3]]) # 3행 1열의 2D
d1.size()
```

```python
# 크기가 1인 차원 삭제
d2=d1.squeeze()
print(d1.dim(), d1.size())
print(d2.dim(), d2.size()) # 차원이 삭제되면서 1차원 벡터가 됨
```

```python
d3=torch.rand(3,2,1)
d4=d3.squeeze()
print(d3.dim(), d3.size())
print(d4.dim(), d4.size())
```

```python
d1=torch.FloatTensor([[1],[2],[3]])
print(d1.shape)
```

```python
# 인덱스 0에다가 1차원 추가
d2=d1.unsqueeze(0) # 인덱스 0에다가 1차원 추가
print(d2.dim(), d2.shape) # [1,3,1]

d3=d1.unsqueeze(1) # 인덱스 1에다가 1차원 추가
print(d3.dim(), d3.shape) # [3,1,1]
```

```python
d1=torch.randn(1,3,1,5)
d2=d1.squeeze(0) # 모든 1인 차원 다 제거됨
print(d1.shape, d2.shape)
d3=d1.squeeze(2)
print(d2.shape)
```
