```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('C:/a/run/hour.csv')
data
```

```python
data.isna().sum()
```

```python
feature=data['weathersit'].values.reshape(-1,1) 
target=data['cnt']
```

```python
X_train, X_test, y_train, y_test=train_test_split(feature, target, test_size=0.2, random_state=42)
```

```python
reg=linear_model.LinearRegression()
reg.fit(X_train, y_train)
```

```python
y_pred=reg.predict(X_test)
r2_score(y_test, y_pred)
```

```python
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test,y_pred, color='red', ls=':')
```

```python
plt.hist(X_test, bins=100, color='red', edgecolor='black')
plt.hist(y_pred, bins=100, color='blue', edgecolor='yellow')
```
