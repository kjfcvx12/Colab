```python
# 데이터 정규화(표준화)를 위해 사용되는 클래스
from sklearn.preprocessing import StandardScaler
```

```python
# StandardScaler 객체 생성
scaler=StandardScaler()
```

```python
# 데이터 정규화하고 변수에 저장
# 데이터의 통계(평균, 표준편차)를 학습한 이후에 학습한 통계를 사용해서 데이터를 변환

# 값의 범위가 너무 크다보면 학습이 잘 안된다..
f_scaled=scaler.fit_transform(features)
```

```python
#numpy 배열로 변환
f_scaled
```

```python
#pandas의 dataframe으로 변환
f_scaled=pd.DataFrame(f_scaled,columns=features.columns)
```

```python
fig, axis=plt.subplots(1,2,figsize=(5,3))

axis[0].hist(f_scaled['Age'], bins=10, color='red')
axis[1].hist(features['Age'], bins=10, color='blue')
```

```python
# 데이터셋을 랜덤하게 섞은 후, 학습용과 검증용으로 분할
# 모델학습-> 검증 -> 테스트
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(f_scaled, label, test_size=0.2, random_state=1)
# X_train : 학습용 문제 데이터(feature) -> 모델 학습용 입력값
# X_test : 테스트용 문제 데이터(feature) -> 학습한 모델을 평가할 때 쓰는 입력값
# y_train : 학습용 정답 데이터(label) -> 모델 학습에 쓰는 정답
# y_test : 테스트용 정답 데이터(label) -> 모델 성능 평가용 정답

# 학습용 데이터로 모델을 훈련하고(train), 테스트용 데이터로 모델을 평가(test)
# 정규화된 입력, 정답데이터를 8:2로 분리
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score

# 객체 생성
dt=DecisionTreeClassifier()

# 학습데이터로 모델 학습시키기
dt.fit(X_train, y_train)
```

```python
# 학습데이터에 대한 예측값 생성
dt_train=dt.predict(X_train)
dt_train
```

```python
# 테스트데이터에 대한 예측값 생성
dt_test=dt.predict(X_test)
dt_test
```

```python
# 학습데이터에 대한 정확도 계산
acc_train=accuracy_score(y_train,dt_train)
acc_train
```

```python
# 테스트데이터에 대한 정확도 계산
acc_train=accuracy_score(y_test, dt_test)
acc_train
```

```python
# 학습데이터에 대한 f1-score 계산
f1_train=f1_score(y_train, dt_train)
f1_train
```

```python
# 테스트데이터에 대한 f1-score 계산
f1_test=f1_score(y_test,dt_test)
f1_test
```

```python
# 학습 / 테스트 데이터의 정확도, f1_score
# 모델이 문제 패턴을 이해한 것이 아니라 -> 답을 통재로 외워버린 상태 (훈련은 거의 100, 테스트는 낮음) #과적합
```

```python
# 교차검증
from sklearn.model_selection import cross_val_score, KFold

# 전체데이터를 나누는 개수 : K
# 학습데이터: K-1
# 검증데이터: 1(항상 1개)
# 데이터를 K개로 나눠서 교차검증(1개는 검증용, 3개는 학습용으로사용함)
kf=KFold(n_splits=4, shuffle=True, random_state=1)

# 1회차 B,C,D로 학습 -> A로 검증
# 2회차 A,C,D로 학습 -> B로 검증
# accuracy, f1-score
#                        평가모델, 입력데이터, 정답, cv=교차검증분할방법, 평가지표
dt_acc_scores=cross_val_score(dt, f_scaled, label, cv=kf, scoring="accuracy")
dt_f1_scores=cross_val_score(dt, f_scaled, label, cv=kf, scoring="f1")

print(dt_acc_scores)
print(dt_acc_scores.mean())


print(dt_f1_scores)
print(dt_f1_scores.mean())
```

```python
import time

def eval_model(model,X_train, X_test, y_train, y_test):
    start_time = time.time()
    
    model.fit(X_train, y_train)
    elapse_time=time.time()-start_time # 이 모델이 학습하는데 얼마나 걸렸나

    y_train_pred=model.predict(X_train) # 학습데이터에 대한 예측값 -> 모델이 이미 공부한 데이터에 대해 얼마나 잘 기억하나 확인
    y_test_pred=model.predict(X_test) # 테스트데이터에 대한 예측값 -> 모델이 처음 보는 데이터에 대해 얼마나 잘 일반화 하는지 확인

    acc_train=accuracy_score(y_train, y_train_pred)
    f1_train=f1_score(y_train, y_train_pred)

    acc_test=accuracy_score(y_test, y_test_pred)
    f1_test=f1_score(y_test, y_test_pred)

    return acc_train, f1_train, acc_test, f1_test, elapse_time
```

```python
model1=GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=1)
# n_estimators : 나무의 개수
# learning_rate : 앞선 나무의 오차를 얼마나 반영할지 결정하는 "보폭"
# max_depth : 나무가 너무 깊으면 각각 데이터에 너무 예민해짐 -> 부스팅에서는 3~5사이의 얕은 나무를 여러개 이어붙이는 방식 선호
# random_state : 코드 다시 돌려도 똑같은 성능과 결과 얻기 위해 설정하는 고정값
```

```python
model1
```

```python
acc_train, f1_train, acc_test, f1_test, elapse_time=eval_model(model1,X_train, X_test, y_train, y_test)
```
